## K-Means++
- K-Means is quite sensitive to initalization, if the init is not good, our algorithm is not able to make desired number of clusters.
- It initialzes the centers for optimal result
- The method is default choice for sklearn

### Using Sci-kit learn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500, n_features=2,centers=5,random_state=3)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=5)
```

```python
km.fit(X,y)
```

```python
centers = km.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.show()
```

```python
pred = km.labels_
```

```python
print(pred)
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black')
plt.show()
```
